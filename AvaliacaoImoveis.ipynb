# Conhecimento do Negócio
- Nesta etapa nosso objetivo é descrever o que já sabemos de antemão sobre o negócio e como isso influencia diretamente em nossa análise.

- A base de dados utilizada para treinar o modelo foi retirada de um levantamento de casos reais de um portal de imóveis do Rio de Janeiro, localizados principalmente nos bairros da Zona Sul, Barra da Tijuca e Recreio.
- Nossos dados não contemplam aluguéis, portanto, as colunas relativas ao aluguel do imóvel serão removidas da base de dados.
- Os valores de condomínio e IPTU não são confiáveis, portanto, vamos remover também essas colunas para não influenciar negativamente no nosso modelo. 
- Os campos de Área Total e Área Útil possuem os mesmo valores, portanto, vamos utilizar apenas o campo Área Total e descartar a Área Útil.
- Nosso Database possui apenas imoveis da cidade do Rio de Janeiro(RJ), portanto, vamos remover as colunas de Cidade e Estado também.
- O campo de Benefícios possui os diversos benefícios do imóvel separados por ";", portanto, vamos ter que primeiro extrair esses valores para depois criar as colunas com eles.
- Certas características do Imóvel estão no campo descrição, portanto, vamos ter que extrair os dados mais frequentes contidos nas descrições para poder analisá-los depois.
- Tipologia dos Imoveis: 
Para fins de agrupamento e tomadas de decisão, os imóveis serão organizados por um tipologia determinada pelo número de quartos e área total em m2, de acordo com as seguintes regras:

  - T0 = Imóveis de 0 quartos com área minima de 35m2
  - T1 = Imóveis de 1 quarto com área minima de 52m2
  - T2 = Imóveis de 2 quartos com área minima de 72m2
  - T3 = Imóveis de 3 quartos com área minima de 91m2
  - T4 = Imóveis de 4 quartos com área minima de 105m2
  - T5 = Imóveis de 5 quartos com área minima de 122m2
  - T6 = Imóveis de 6 quartos com área minima de 134m2

  Referencia da Tipologia: 

  Utilizaremos como base o modelo de tipologia de imóveis adotado em Portugal, pois se trata de uma separação adequada para iniciarmos nossos estudos.

## Criando as novas colunas

Desmembrando o campo Beneficios em campos separados no DataFrame

Conhecendo o campo Beneficios

```python
# Verifica o cabecalho somente no campo Beneficios para verificar o formato 
imoveis['Beneficios'].head(3)
```

Extraindo as características de dentro do campo benefícios

```python
# Cria uma copia temporaria do dataframe imoveis para manipulacao
imoveiscopia = imoveis

# Splita os Beneficios pelo separador e cria um novo dataframe
beneficios = imoveiscopia['Beneficios'].str.split(',', expand=True)

# Dropa os valores NaN
beneficios = beneficios.dropna(axis=1)

# Cria um array para os beneficios unicos independente da linha/coluna onde está
beneficios_unicos = []
    
# Varre todas as linhas e colunas alimentando o array de beneficios unicos
for index, row in beneficios.iterrows():
  columns = beneficios.columns.values
  for col in columns:        
    column_name = row[col]    
    beneficios_unicos.append(column_name)

# remove as duplicatas do array
beneficios_unicos = list(dict.fromkeys(beneficios_unicos))

# remove os vazios
beneficios_unicos.remove('')

# visualiza a lista de beneficios unicos
print(beneficios_unicos)
```

Criando e preenchendo os novos campos no Dataframe, de acordo com os benefícios de cada imóvel

```python
# gera um array com as colunas do dataframe imoveiscopia
array_colunas = imoveiscopia.columns

# varre o array de beneficios unicos verificando se existe a coluna no array de colunas
# se nao existe a coluna no array, cria ela no dataframe 
for beneficio in beneficios_unicos:
  existe = beneficio in array_colunas
  if not existe:
    imoveiscopia[beneficio] = 0

# copia de volta o dataframe temporario com as colunas novas por cima dos imoveis
imoveis = imoveiscopia

# visualiza o cabecalho pra ver como ficou
imoveis.head(3)
```

## Preenchendo as novas colunas


Preenchendo os campos novos com valores 0 (Não Possui) ou 1 (Possui) de acordo com a presença da características no campo de Benefícios

```python
# vamos novamente manipular a copia para nao afetar os dados originais caso algo dê errado
imoveiscopia = imoveis

# para cada beneficio filtra os imoveis que contem a caracteristica no campo de beneficios dos imoveis
# para cada imovel preenche o campo relacionado com 1 (Possui)
for beneficio in beneficios_unicos:    
  imoveisfiltro = imoveiscopia[imoveiscopia['Beneficios'].str.contains(beneficio)]
  imoveiscopia.loc[imoveiscopia.index.isin(imoveisfiltro.index), [beneficio]] = 1

# retorna o backup pro dataframe de imoveis
imoveis = imoveiscopia

# visualiza o cabecalho do dataframe para conferir o resultado
imoveis.head(3)


```

#### Discrete joint distribution

Let’s see how the cars in our data set are distributed according to horsepower and drivetrain layout. That is, we want to visualize the following table

|  horsepower ↓ \ drive-wheels → | 4wd | fwd | rwd |
|-------------------------------|-----|-----|-----|
|  Low (0-100)                  | 5   | 89  | 15  |
| Medium (100-150)              | 3   | 24  | 35  |
| High (150+)                   | 0   | 5   | 25  |


First we calculate the counts per groups, to get values corresponding to the table above


```python
imoveis['cnt'] = np.ones(len(imoveis))
bin_labels = ['Pequenos (0-100)', 'Medios (100-150)', 'Grandes (150+)']

imoveis['area-group'] = pd.cut(imoveis['Area_Total'], [0, 100, 150, imoveis['Area_Total'].max()], labels=bin_labels)
g = imoveis.groupby(['area-group', 'Quartos']).count()[['cnt']].replace(np.nan, 0).reset_index()
#g = g.loc[g['Quartos'] <= 3]
display(g)
```

Then we can use the `heatmap` function to create a heatmap representing our distribution.

Let's use the `size` parameter. It will enable us to map item counts (ie. distribution frequences) to size in order to help our eyes see the distribution more clearly.

```python
plt.figure(figsize=(5, 5))

heatmap(
    x=g['Quartos'], # Column to use as horizontal dimension 
    y=g['area-group'], # Column to use as vertical dimension
    
    size_scale=7000, # Change this to see how it affects the plot
    size=g['cnt'] # Values to map to size, here we use number of items in each bucket
    
)
```

Note that we used the same series for color and size, which is what we usually want to do.

### Definindo a Baseline

**Definindo uma Baseline**

- `Uma baseline é importante para ter marcos no projeto`.
- `Permite uma explicação fácil para todos os envolvidos`.
- `É algo que sempre tentaremos ganhar na medida do possível`.

```python
# agrupa as categorias e calcula as médias
medias_categorias = imoveis.groupby(by='Tipologia_Area')['ALVO'].mean()
```

```python
# visualizando a variável medias_categorias
medias_categorias
```

```python
# criando o dicionario com chaves e seus valores
dic_baseline = {'Pequeno': medias_categorias[150],  'Medio': medias_categorias[300],  'Grande': medias_categorias[400]}
```

```python
# imprime dicionario
dic_baseline
```

```python
# cria a função retorna baseline
def retorna_baseline(area):
  #if area <= 100:
  #  return dic_baseline.get('Pequeno-100')
  if area <= 150:
    return dic_baseline.get('Pequeno')
  #elif area <= 200:
  #  return dic_baseline.get('Medio-200')
  #elif area <= 250:
  #  return dic_baseline.get('Medio-250')
  elif area <= 300:
    return dic_baseline.get('Medio')
  #elif area <= 350:
  #  return dic_baseline.get('Grande-350')
  elif area <= 400:
    return dic_baseline.get('Grande')
  else:
    return dic_baseline.get('Enorme')
```

### Tunning do Algoritmo de Machine Learning

**Parâmetros da Random Forest**

- `O parâmetro n_estimators define a quantidade de árvores que serão usadas para criação dos modelos. Quanto maior o número de árvores, maior será o número de modelos no ensemble.`

- `O parâmtro max_deph define a profundidade da árvore, quanto mais profunda a árvore, mais especializado nos dados de treino o nosso modelo estará.`

- `O parâmetro max_features define a quantidade máxima de nós que serão usados para separar os níveis folha da árvore`
